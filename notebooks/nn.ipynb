```python
%%mojo draw_neuron
from python import PythonObject
from unnet import Neuron, Layer, NetworkMLP
from unnet.utils import draw

fn draw_neuron() raises -> PythonObject:
    neuron = Neuron(weights=[0.7, 0.8], bias=0.5)
    result = neuron([2.0, 3.0])
    result.backward()
    return draw(result)
```

```python
%%mojo
from unnet import Neuron, Layer, NetworkMLP
from unnet.utils import draw

fn main() raises:
    # training a neural network to choose the minimum of three numbers
    xs = [
        [3.0, 2.0, -1.0],
        [1.0, 0.0, -0.5],
        [0.5, 2.0, 1.0],
        [1.0, 1.0, 0.0],
        [1.0, 2.0, 3.0],
        [0.0, 1.0, 0.0],
    ]  # inputs
    ys = [[-1.0], [-0.5], [0.5], [0.0], [1.0], [0.0]]  # targets (minimum number of each input)
    # neural network with 3 inputs, 3 hidden neurons, and 1 output neuron
    network = NetworkMLP.create_random(2, 3, 3, 1)
    
    loss_list = network.train(training_data=xs, desired_output=ys, steps=1000)
    for i, loss in enumerate(loss_list):
        if i % 10 == 0:
            print(loss)  # print loss for each 10 steps
```
